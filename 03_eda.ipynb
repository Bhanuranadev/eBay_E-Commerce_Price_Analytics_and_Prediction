```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/cleaned_data.csv")

print("Shape:", df.shape)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=30, kde=True)
plt.title("Price Distribution of eBay Products")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

#Insights:-Most products are priced in the lower to mid price range, with a few high-priced outliers.

```python
top_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Product Categories on eBay")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.show()
```

Insight:Certain categories dominate product listings, indicating higher supply in those segments.

```python
avg_price_category = (
    df.groupby("category")["price"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_price_category.values, y=avg_price_category.index)
plt.title("Top 10 Categories by Average Price")
plt.xlabel("Average Price")
plt.ylabel("Category")
plt.show()
```

Insight:

Categories such as electronics tend to have higher average prices compared to others.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="condition", y="price", data=df)
plt.title("Price vs Product Condition")
plt.xlabel("Condition")
plt.ylabel("Price")
plt.show()
```

Insight:

New or better-condition products generally command higher prices.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="currency", data=df)
plt.title("Currency Distribution")
plt.show()
```

Insight:

Most products are listed in a single dominant currency.
